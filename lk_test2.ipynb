```python
from google.colab import files
import os
from lightkurve import TessTargetPixelFile

# Step 1: Upload the file
uploaded = files.upload()

# Step 2: Verify the upload
print(os.listdir())
```

```python
# Step 3: Load the uploaded file
tpf = TessTargetPixelFile("tess2022302161335-s0058-0000000327507922-0247-s_tp.fits")
tpf
tpf.plot(aperture_mask=tpf.pipeline_mask)

lc = tpf.to_lightcurve()
lc.scatter()
flat_lc = lc.flatten()
flat_lc.scatter(s=1)

import numpy as np
periodogram = flat_lc.to_periodogram(method="bls", period=np.arange(1, 16, 0.001))
periodogram.plot()

best_fit_period = periodogram.period_at_max_power
print('Best fit period: {:.3f}'.format(best_fit_period))
```
